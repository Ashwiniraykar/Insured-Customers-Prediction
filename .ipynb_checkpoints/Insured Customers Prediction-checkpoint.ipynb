# Predicting the Amount of Money Spent on Insured Customers 

### Analyzing insured customers' data for an insurance company:
**To predict the dollar amount of money spent by the insurance company on insured customers, based on a sample data that consists of the profile of insured customers**

# Insured customers' Data
The insured customers' data is in a csv file. It has information sconsisting of:
1. age
2. sex (female, male)
3. BMI
4. Children
5. Smoker (yes, no)
6. Region (northeast, northwest, southeast, southwest])
7. **expenses**

The value we want to predict is **expenses**


```python
import numpy as  np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("insurance Dataset.csv")
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.info()
```

```python
#Convert the columns with text data into numeric data
df["female"] = df["sex"]
df["female"].replace(["female","male"],[1,0],inplace=True)
df["male"] = df["sex"]
df["male"].replace(["female","male"],[0,1],inplace=True)
df["yes"] = df["smoker"]
df["yes"].replace(["yes","no"],[1,0],inplace=True)
df["no"] = df["smoker"]
df["no"].replace(["yes","no"],[0,1],inplace=True)
df["southwest"] = df["region"]
df["southwest"].replace(["southwest","southeast","northwest","northeast"],[1,0,0,0],inplace=True)
df["southeast"] = df["region"]
df["southeast"].replace(["southwest","southeast","northwest","northeast"],[0,1,0,0],inplace=True)
df["northwest"] = df["region"]
df["northwest"].replace(["southwest","southeast","northwest","northeast"],[0,0,1,0],inplace=True)
df["northeast"] = df["region"]
df["northeast"].replace(["southwest","southeast","northwest","northeast"],[0,0,0,1],inplace=True)
df
```

```python
#Expenses are in proportion with smoker and age. 
#As per the data,if male or female smoking then thier expenses are more.
#likewise, if age is more then also their expenses are more.
```

```python
# Refer to the customers features data by "X", and refer to the label feature (expenses) by "y" 
y = df['expenses']
x = df[['age','bmi','children','female','male','yes','no','southwest','southeast','northwest','northeast']]
print(y, '\n')
print(x)
```

```python
# Load the train_test_split function
from sklearn.model_selection import train_test_split
```

```python
# Split the data into:
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=49)
print("x_train=",x_train)
print("y_train=",y_train)
print("x_test=",x_test)
print("x_test=",x_test)
```

```python
#Import LinearRegression from sklearn.linear_model and create an instance of a LinearRegression() model named
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

## 1. linear model's intercept and coefficients

```python
# linear model's intercept and coefficients
print('Intercept', lm.intercept_)
print('Coefficient', lm.coef_)
```

```python
lm.fit(x_train,y_train)
```

```python
# trained model to predict the test data set
predictions = lm.predict(x_test)
predictions
```

```python
#plt.scatter(yprint('correlation coefficients:\n', np.corrcoef(x, y))_test, predictions)
#print('correlation coefficients:\n', np.corrcoef(x_test, y_test))
plt.scatter(y_test, predictions)
plt.xlabel('test value')
plt.ylabel('Predicted value')
```

```python
#1. Mean Absolute Error
#2. Mean Squared Error
#3. Root Mean Squared Error.

from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

## 2. Implementation of the Gradient Descent to find the intercept and coeficients of the linear regression model


```python
# implementation of the Gradient Descent to find the intercept and coeficients of the linear regression model
X_b = np.c_[np.ones((1338, 1)), x]
alpha = 0.00001
n_iterations = 100000
m = 1338
W = np.random.randn(12,1)
y1=y.values
y2=y1.reshape((1338,1))
for iteration in range(n_iterations):
    gradients = 1/m * X_b.T.dot(X_b.dot(W) - y2)
    W = W - alpha * gradients
print(W)
y2.shape
```
